```python
import pandas as pd
df = pd.read_csv('salaries_by_college_major.csv')
clean_df = df.dropna()
```

```python
# shows the difference between the highest and lowest earners in a career
clean_df['Mid-Career 90th Percentile Salary'] - clean_df['Mid-Career 10th Percentile Salary']
```

```python
# does the same as above
clean_df['Mid-Career 90th Percentile Salary'].subtract(clean_df['Mid-Career 10th Percentile Salary'])
```

```python
# we can add this to the existing dataframe. The '1' is the position, so the second column
spread_col = clean_df['Mid-Career 90th Percentile Salary'] - clean_df['Mid-Career 10th Percentile Salary']
clean_df.insert(1, 'Spread', spread_col)
clean_df.head()
```

```python
# finding the careers with the lowest spread
low_risk = clean_df.sort_values('Spread')
low_risk[['Undergraduate Major', 'Spread']].head()
```

```python
# challenge to find the top 5 degrees with the highest values in the 90th percentile
top_earners = clean_df.sort_values(ascending=False, by='Mid-Career 90th Percentile Salary')
top_earners[['Undergraduate Major', 'Mid-Career 90th Percentile Salary']].head()
```

```python
# challenge to find the degrees with the highest spread in salaries
high_risk = clean_df.sort_values(ascending=False, by='Spread')
high_risk[['Undergraduate Major', 'Spread']].head()
```
